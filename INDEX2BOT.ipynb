```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
import pandas as pd
import os
import time
from urllib.parse import quote

# Configurar o serviço do ChromeDriver
servico = Service(ChromeDriverManager().install())
navegador = webdriver.Chrome(service=servico)

# Definir o caminho para o arquivo Excel
database = os.path.join('C:\\Users\\Usuário\\Curso_Python\\Prototipo_zapBot\\pasta', 'telefones6.xlsx')

# Carregar a tabela de números de telefone
tabela = pd.read_excel(database)
lista = tabela['Telefone']

# Abrir o WhatsApp Web uma vez no início
navegador.get("https://web.whatsapp.com/")
# Aguardar o tempo necessário para o usuário escanear o QR code
WebDriverWait(navegador, 60).until(EC.presence_of_element_located((By.ID, "side")))

mensagem = 'Olá! Você sabia que sua startup pode conseguir recursos significativos sem precisar vender parte da empresa e sem fazer empréstimo? Conseguimos te ajudar a realizar seu projeto inovador. Vamos conversar sobre como isso é possível e como sua empresa pode se beneficiar dessa oportunidade. Vamos conversar?'

# Loop para enviar mensagens
for telefone in lista:
    try:
        # Formatar o número de telefone para o link do WhatsApp
        telefone_formatado = quote(f"+{telefone}")
        link = f"https://web.whatsapp.com/send?phone={telefone_formatado}&text={quote(mensagem)}"
        
        navegador.get(link)
        
        # Esperar que o campo de mensagem esteja presente
        campo_mensagem = WebDriverWait(navegador, 30).until(
            EC.presence_of_element_located((By.XPATH, "//div[@contenteditable='true'][@data-tab='10']"))
        )
        time.sleep(10)
        # Aguardar o botão de envio estar clicável e enviar a mensagem
        botao_enviar = WebDriverWait(navegador, 30).until(
            EC.element_to_be_clickable((By.XPATH, "//span[@data-icon='send']"))
        )
        time.sleep(7)
        # Enviar a mensagem
        campo_mensagem.send_keys(Keys.ENTER)
        botao_enviar.click()
        
        # Aguardar um momento antes de enviar a próxima mensagem
        time.sleep(5)
    except Exception as e:
        print(f'Não foi possível enviar mensagem para {telefone}: {e}')

# Fechar o navegador ao final
```
